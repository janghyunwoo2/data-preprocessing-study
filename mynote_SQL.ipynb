# 3장 집약

#### 집약이란?
데이터의 가치를 되도록 손실 없이 압축하여 데이터의 단위(데이터 행의 의미)를 변환할 수 있도록 하는 처리를 의미

### 3.1 데이터와 종류의 개수 산출

GROUP_BY 로 집약 단위 지정
COUNT나 distinct 사용해서 집계 처리

```python
# 데이터 집계 : 대상 데이터의 행수를 세는 처리
cur.execute('''
SELECT
  -- 집약 단위인 호텔 ID 추출
  hotel_id,

  -- COUNT 함수에 reserve_id를 지정하고 있기 때문에, reserve_id NULL이 아닌 행을 카운트
  COUNT(reserve_id) AS rsv_cnt,

  -- customer_id에 distinct를 설정하여 중복을 제거
  -- 중복을 제거한 customer_id의 수를 센다
  COUNT(distinct customer_id) AS cus_cnt

FROM reserve_tb

-- GROUP BY 구문에서 지약할 단우를 hotel_id로 지정
GROUP BY hotel_id
''') 
rows = cur.fetchall()

pandas_view(rows)
```

### 3.2 합곗값 계산

월별 매출을 알아야 하거나 등 분석 대상의 값이 숫자일 때 데이터의 합을 계산

```python
# 합곗값 산출
cur.execute('''
SELECT
  hotel_id,
  people_num,

  -- SUM함수에 total_price를 지정하여 매출 합계 금액을 산출
  SUM(total_price) AS price_sum

FROM reserve_tb

-- 집약 단위를 hotel_id와 people_num의 조합으로 지정
GROUP BY hotel_id, people_num
''') 
rows = cur.fetchall()

pandas_view(rows)
```

### 3.3 최댓값, 최솟값, 대푯값 산출

* 대푯값 = 중앙값 = 50% 백분위

SQLite에는 median,PERCENTILE_CONT 함수 없다.

```python
# 합곗값 산출
cur.execute('''
SELECT
  hotel_id,

  -- total_price의 최대값 산출
  MAX(total_price) AS price_max,

  -- total_price의 최소값 산출
  MIN(total_price) AS price_min,

  -- total_price의 평균값 산출
  AVG(total_price) AS price_avg

  -- total_price의 중앙값 산출
  -- MEDIAN(total_price) AS price_med   

  -- PERCENTILE_CONT 함수에 0.2를 지정하여 20% 백분위를 산출
  -- ORDER BY 구문에 total_price를 지정하여 백분위 값의 대상 열과 데이터의 정렬 방식을 지정
  -- PERCENTILE_CONT(0.2) WITHIN GROUP(ORDER BY total_price) AS price_20per

FROM reserve_tb
GROUP BY hotel_id
''') 
rows = cur.fetchall()

pandas_view(rows)
```

### 3.4 분포 계산

분산값과 표준편찻값은 데이터 분포 정도를 나타낸다.

데이터의 전체적인 경향을 더욱 잘 표현 할 수 있다.

##### 주의 할점 
분산값과 표준편찻값의 계산식에서 데이터수-1을 이용해 나누는 부분이 있다
이때 데이터 수가 1개이면 1-1=0으로 나누어 잘못된 값이 나온다
데이터 1가 1개라는 것은 분포의 의미가 없으므로 분산값 표준편찻값 모두 0으로 만들어준다.

##### SQLite에서 분산값 Variance, 표준편차 : STDDEV 지원 안함.


```python
# 분산값 : Variance, 표준편차 : STDDEV
cur.execute('''
SELECT
  hotel_id,

  -- VARIANCE 함수에 total_price를 지정하여 분산값 산출
  -- COALESCE 함수로 분산값이 NULL일 때 0으로 변환
  COALESCE(VARIANCE(total_price), 0) AS price_var,

  -- 데이터 수가 2건 이상이면 STDDEV 함수에 total_price를 지정하여 표준편차 값을 산출
  COALESCE(STDDEV(total_price), 0) AS price_std

FROM reserve_tb
GROUP BY hotel_id
''') 
rows = cur.fetchall()

pandas_view(rows)
```

### 3.5 반올림(최빈값 구하는데 사용)

나이같은 연속적인 값을 10대 20대 처럼 범주형으로 바꾸기위해

반올림 내림 올림 등으로 처리해주는 것

ROUND 함수: 반올림 하는 함수  

* 예를들어 356.44 경우 

0으로 설정하면 소수점 0.1자리에서 반올림 => 356

1로 설정하면 0.01자리에서 반올림 => 356.4

-1로 설정하면 1의 자리에서 반올림 => 360

```python
# 최빈값: 자주 등장하는 값 구하기
cur.execute('''
SELECT
  ROUND(total_price, -2) AS total_price_round,COUNT(*)
FROM reserve_tb
GROUP BY total_price_round

-- COUNT 함수로 계산한 금액별 예약 수를 큰 순서대로 정렬(DESC 는 내림차순)
-- count(*) 은 null값을 포함한 행 전체를 카운트, 여기선 total_price_round으로 바꿔도 똑같은 결과
ORDER BY COUNT(*) DESC

-- LIMIT 구문으로 첫 번째 결과 값만 얻음
-- LIMIT 1
''') 
rows = cur.fetchall()

pandas_view(rows)
```

### 3.6 순위 계산

그룹별로 순서를 정렬하고 순위를 매기는 계산은 Window 함수를 이용하면 간략하고 성능 좋은 코드 작성 가능

Window 함수는 집약 함수 중 하나인데, 일반적인 집약함수와 비교하여 행을 집약하는 것이 아니라 집약한 결과를 계산하여 각 행에 첨부하는 점이 다릅니다.

```python
cur.execute('''
SELECT
  customer_id,reserve_datetime,
  -- ROW_NUMBER로 순위를 구함
  -- PARTITION by customer_id로 고객 별 순위를 구하도록 설정
  -- ORDER BY reserve_datetime으로 오래된 예약 시간 순으로 순위를 구함
  ROW_NUMBER() OVER (PARTITION BY customer_id ORDER BY reserve_datetime) AS log_no

    FROM reserve_tb
''') 
rows = cur.fetchall()

pandas_view(rows)
```

### 3.7 랭킹

Rank 함수 이용

위 Row_number과 차이는 

row_number은 같은 값이라도 순서에 따라 순위가 달라지는 반면(예: 같은 80점이라도 뒤에오는 것이 순위가 낮음)

Rank 함수는 같은 값이면 같은 순위로 정해준다(예: 같은 80점 이면 같은 등수 임)

```python
cur.execute('''
SELECT
  hotel_id,

  -- RANK 함수로 예약수 순위를 지정 
  -- COUNT(*)를 RANK의 기준으로 지정(집약 후의 예약수에 따라 순위를 정하는 계산)
  -- DESC로 내림 차순으로 정렬
  RANK() OVER (ORDER BY COUNT(hotel_id) DESC) AS rsv_cnt_rank

FROM reserve_tb

-- hotel_id를 집약 단위로 지정, 예약수를 계산하기 위한 설정으로 RANK 함수와는 무관
GROUP BY hotel_id
''') 
rows = cur.fetchall()

pandas_view(rows)
```

# 4장 결합

#### 결합이란?
데이터를 분석하기 위해선 하나의 테이블에 모두 정리된 테이블이 이상적

그것을 join을 이용하며 처리한다.


### 4.1 조인

join 함수를 이용해서 결합

### 주의!!
join은 모든 처리 중 가장 계산량이 많아지기 쉽다.

그래서 테이블을 결합하고 조건문으로 줄이는 것 보다

미리 테이블을 조건으로 줄여 놓고 결합하는게 효과적이다.

```python
# 결합하기전 미리 조건으로 줄여준다. on 뒤에는 where처럼 조건 붙일수 있다.
cur.execute('''
SELECT
  -- 필요한 열 추출
  rsv.reserve_id, rsv.hotel_id, rsv.customer_id,
  rsv.reserve_datetime, rsv.checkin_date, rsv.checkin_time, rsv.checkout_date,
  rsv.people_num, rsv.total_price,
  hotel.base_price, hotel.big_area_name, hotel.small_area_name,
  hotel.hotel_latitude, hotel.hotel_longitude, hotel.is_business

FROM reserve_tb rsv
  JOIN hotel_tb hotel
  ON rsv.hotel_id = hotel.hotel_id

    -- 호텔 데이블에서 비지니스 호텔 데이터만을 추출
    AND hotel.is_business = 'True'

-- 예약 테이블에서 비지니스 호텔의 데이터만을 추출
AND rsv.people_num = 1

''') 
rows = cur.fetchall()

pandas_view(rows)
```

### 4.2 조건에 따라 테이블 변경하기

4.1의 연장선 조인하기 전에 결합할 테이블을 줄이는게 중간테이블을 작게해서 좀 더 효율적으로 동작하게 한다.

#### 복잡해서 일단 넘어감
#### 여기서 알아야 하건 with로 임시 테이블를 만들어 사용할 데이터를 작게 만든다.

```python
# 결합하기전 미리 조건으로 줄여준다. on 뒤에는 where처럼 조건 붙일수 있다.
cur.execute('''
-- small_area_name 별로 호텔 수를 계산, 결합 키를 판정하기 위한 테이블
WITH small_area_mst AS(
  SELECT
    small_area_name,

    -- 20건 이상이면 join_area_id에 small_area_name을 지정
    -- 20건 미만이면 join_area_id에 big_area_name을 지정
    -- -1로 자신을 제외
    CASE WHEN COUNT(hotel_id)-1 >= 20
			THEN small_area_name ELSE big_area_name END AS join_area_id

  FROM hotel_tb
  GROUP BY big_area_name, small_area_name
)
-- recommend_hotel_mst는 추천 후보를 위한 테이블
, recommend_hotel_mst AS(
  -- join_area_id를 big_area_name으로 한 추천 후보 마스터
  SELECT
    big_area_name AS join_area_id,
    hotel_id AS rec_hotel_id
  FROM hotel_tb

  -- union으로 테이블을 연결
  UNION

  -- join_area_id을 small_area_name으로 한 추천 후보 마스터
  SELECT
    small_area_name AS join_area_id,
    hotel_id AS rec_hotel_id
  FROM hotel_tb
)
SELECT
  hotels.hotel_id,
  r_hotel_mst.rec_hotel_id

-- 추천 대상의 hotel_tb 읽기
FROM hotel_tb hotels

-- 각 호텔의 추천 후보 대상 지역을 판단하기 위해 small_area_mst를 결합
INNER JOIN small_area_mst s_area_mst
  ON hotels.small_area_name = s_area_mst.small_area_name

-- 대상 지역의 추천 후보 결합
INNER JOIN recommend_hotel_mst r_hotel_mst
  ON s_area_mst.join_area_id = r_hotel_mst.join_area_id

  -- 추천 후보에서 자신을 제외
  AND hotels.hotel_id != r_hotel_mst.rec_hotel_id


''') 
rows = cur.fetchall()

pandas_view(rows)
```

### 4.3 과거 데이터에서 정보 얻기

호텔 예약을 예측 할 때 과거 예약한 정보를 이용하면 유용할 것이다.

그래서 현재와 과거 데이터를 결합해서 새로운 행을 만든다.

하지만 무턱대고 한 최신 데이터에 모든 과거 데이터를 결합 시키는 것은 너무 방대해 지므로 비추천

그래서 두가지 방법으로 결합할 과거 데이터를 줄인다.

1. 결합 대상이 대는 과거의 기간을 줄인다.
2. 결합한 과거 데이터에 집약 함수를 사용하여 데이터 수가 늘어나지 않도록 한다.(예: 과거 데이터의 평균을 내서 결합시키는 방법. 이방법은 과거 데이터가 5년이던 10년이던 평균으로 바꾸기 때문에 결국 최신 데이터에 결합되는 데이터는 하나 이다.)


#### Q1 n건 전의 데이터 얻기

```python
cur.execute('''
SELECT
	*,
    -- join 함수로 하는 것보다 lag 함수가 효율적. lag함수는 window 함수중 하나로 n건
	-- LAG 함수를 이용하여 2건 전의 total_price를 before_price로 구함
	-- LAG 함수로 참조할 그룹을 customer_id로 지정
	-- LAG 함수로 참조할 그룹의 데이터를 reserve_datetime이 오래된 순서로 정렬
  LAG(total_price, 2) OVER
	(PARTITION BY customer_id ORDER BY reserve_datetime) AS before_price

FROM reserve_tb

''') 
rows = cur.fetchall()

pandas_view(rows)
```

#### Q2 과거 n건의 합곗값

Q1는 단지 과거의 소비 한개만 구했는데 Q2는 과거 소비 n건의 합곗값을 구해 데이터 정확도를 높히려 함.(평균을 구하려고 하는 것이니 단편적인 하나의 값과 비교하는 것 보단 정확할 것이니..)

```python
cur.execute('''
SELECT
  *,
  CASE WHEN

    -- COUNT 함수로 합계를 계산한 건수를 세서 3건인지 확인
		-- BETWEEN 구문으로 2건 전부터 
  	COUNT(total_price) OVER
		(PARTITION BY customer_id ORDER BY reserve_datetime ROWS
		 BETWEEN 2 PRECEDING AND CURRENT ROW) = 3

  THEN

    -- 자신을 포함한 3건의 합계 금액을 계산
  	SUM(total_price) OVER
		(PARTITION BY customer_id ORDER BY reserve_datetime ROWS
		 BETWEEN 2  PRECEDING AND CURRENT ROW)

  ELSE NULL END AS price_sum

FROM reserve_tb

''') 
rows = cur.fetchall()

pandas_view(rows)
```

#### Q3 과거 n건의 평균값

자신의 행을 포함하지 않고 한 건 이전의 데이터에서 세 건 이전까지의 평균 예약 금액을 첨부하고, 과거 예약 건수가 세 건 미만이면 존재하는 건수의 평균값을 계산한다.

예약이 한 건도 없으면 값을 가지지 않는다.

윈도우 함수인 AVG 함수를 사용하면 된다.

```python
cur.execute('''
SELECT
  *,

  AVG(total_price) OVER
	(PARTITION BY customer_id ORDER BY checkin_date ROWS
	 BETWEEN 3 PRECEDING AND 1 PRECEDING) AS price_avg

FROM reserve_tb

''') 
rows = cur.fetchall()

pandas_view(rows)
```

#### Q4 과거 n일(day)의 합곗값

자신을 포함하지 않으면서 같은 고객의 지난 90일간의 합계 예약 금액 정보를 첨부한다.

```python
cur.execute('''
SELECT
	-- 결합할 모든 데이터열 얻기 
  base.*,

	-- 대상 건수가 0이면 0, 1건 이상 있으면 합계 금액을 계산
  COALESCE(SUM(combine.total_price), 0) AS price_sum

-- 결합할 예약 테이블 지정
FROM reserve_tb base

-- 과거 정보로 결합할 예약 테이블 지정
LEFT JOIN reserve_tb combine

	-- 같은 customer_id 로 결합
  ON base.customer_id = combine.customer_id

	-- 과거 데이터만을 결합 대상으로 지정
  AND base.reserve_datetime > combine.reserve_datetime

	-- 과거 90일 이전까지의 데이터만을 결합 대상으로 지정 ("10장 일시형"에서 자세히 설명)
  AND DATEADD(day, -90, base.reserve_datetime) <= combine.reserve_datetime

-- 결합할 예약 테이블의 모든 데이터열로 집약 
GROUP BY base.reserve_id, base.hotel_id, base.customer_id,
  base.reserve_datetime, base.checkin_date, base.checkin_time, base.checkout_date,
  base.people_num, base.total_price


''') 
rows = cur.fetchall()

pandas_view(rows)
# 파이썬 으로 짠 코드는 성공했으니 결과는 그걸 보자
```

### 4.4 상호 결합

말이 어렵게 나와서 그렇지 그냥 조인에서 키를 지정하지 않고 테이블을 결합하는 것을 말함.

상호 결합은 cross join 구문으로 한다.

```python
cur.execute('''
SELECT
	cus.customer_id,

	-- 연월 마스터 테이블에서 연을 구함
	mst.year_num,

	-- 연월 마스터에서 달을 구함
	mst.month_num,

	-- 해당 total_price가 존재하면 더하고, 없으면 0을 더한다
	SUM(COALESCE(rsv.total_price, 0)) AS total_price_month

FROM customer_tb cus

-- 고객 테이블과 연월 마스터를 상호 결합
CROSS JOIN month_mst mst

-- 고객 테이블과 연월 마스터와 예약 테이블을 결합
LEFT JOIN reserve_tb rsv
  ON cus.customer_id = rsv.customer_id
    AND mst.month_first_day <= rsv.checkin_date
    AND mst.month_last_day >= rsv.checkin_date

-- 연월 마스터의 기간 지정
WHERE mst.month_first_day >= '2017-01-01'
  AND mst.month_first_day < '2017-04-01'
GROUP BY cus.customer_id, mst.year_num, mst.month_num


''') 
rows = cur.fetchall()

pandas_view(rows)
# 파이썬 으로 짠 코드는 성공했으니 결과는 그걸 보자
```

### Q.1 가로 데이터 변환

데이터 셋은 호텔 예약 레코드를 사용한다.

SQL에는 가로 데이터로 변환하기 위한 함수가 제공 되지 않아 CASE 문을 여러번 작성하여 가로 데이터로 변환 해야 된다.

```python
cur.execute('''
WITH cnt_tb AS(
  SELECT
    customer_id, people_num,
    COUNT(reserve_id) AS rsv_cnt
  FROM reserve_tb
  GROUP BY customer_id, people_num
)
SELECT
  customer_id,
  max(CASE people_num WHEN 1 THEN rsv_cnt ELSE 0 END) AS people_num_1,
  max(CASE people_num WHEN 2 THEN rsv_cnt ELSE 0 END) AS people_num_2,
  max(CASE people_num WHEN 3 THEN rsv_cnt ELSE 0 END) AS people_num_3,
  max(CASE people_num WHEN 4 THEN rsv_cnt ELSE 0 END) AS people_num_4
FROM cnt_tb
GROUP BY customer_id


''') 
rows = cur.fetchall()

pandas_view(rows)
# 파이썬 으로 짠 코드는 성공했으니 결과는 그걸 보자
```
